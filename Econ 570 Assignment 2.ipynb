```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import graphviz as gr
```

# Case 1: Oberserved covariate -- control for covariate


```python
g = gr.Digraph()

g.edge("T","Y")
g.edge("X","Y")

g
```

```python
a = 0.5
b = 2
c = 3
n = 100
```

```python

slope_estimates = []
bias = []
rmse = []

for mc_replication in range(1000):
    T = np.random.binomial(1,0.5,n)
    X = np.random.normal(0,1,n)
    e = np.random.normal(0,1,n)
    Y = a+b*T+c*X+e
    exog = pd.DataFrame({"a":T,"b":X})
    mod = sm.OLS(Y,sm.add_constant(exog))
    res = mod.fit()
    slope_estimates = slope_estimates + [res.params[1]]
    bias = bias + [res.params[1]-b]
    rmse = rmse + [(res.params[1]-b)**2]
    
Bias = np.mean(bias)
RMSE = np.sqrt(np.mean(rmse))

    

print(f"estimated treatment effect is: {np.mean(slope_estimates)}")
print(f"bias is: {Bias}")
print(f"rmse is: {RMSE}")

plt.hist(slope_estimates)
```
